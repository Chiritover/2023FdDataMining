```python
original_path = './answer_original.json'
with open(original_path, 'r') as f:
    a = f.readlines()
print(len(a))

    
```

```python
# 留下所有answer，为3i+1行
a = [a[3*i+1] for i in range(0,int(len(a)/3))]
len(a)
```

```python
print(a[0])
```

```python
import re
a =[a.split(':')[1] for a in a]
len(a)
```

```python
print(a[0])
```

```python
import openai
openai_key = os.getenv('OPENAI_API_KEY')
openai.api_key = openai_key
import time
```

```python
yes_count = 0
no_count = 0
no_with_recognize = 0
for i in a:
    # check if Yes is a substring of the string
    if re.search('Yes', i):
        yes_count += 1
    factors = re.findall(r'\d+', i)
    if len(factors) == 1:
        yes_count += 1
        continue
    no_count += 1
    factor  = [int(factors[a]) for a in range(len(factors))]
    print(factor)
    if len(factor) == 0: 
        continue
    original_number = factor[0]
    factor = [factor for factor in factor[1:] if factor != 1 and factor != original_number]
    factor_ans = []
    for factor in factor:
        try:
            completion = openai.ChatCompletion.create(
                model = 'gpt-3.5-turbo',
                messages = [{'role':'user','content':'Is ' + str(original_number) + ' divisible by ' + str(factor) + '? Answer with either Yes or No.'}],
            )
        except:
            time.sleep(60)
            completion = openai.ChatCompletion.create(
                model = 'gpt-3.5-turbo',
                messages = [{'role':'user','content':'Is ' + str(original_number) + ' divisible by ' + str(factor) + '? Answer with either Yes or No.'}],
            )
        answer = completion.choices[0].message['content']
        factor_ans.append(answer)
        with open('factors_question.json', 'a+') as f:
            query = {
                'question': 'Is ' + str(original_number) + ' divisible by ' + str(factor) + '? Answer with either Yes or No.',
                'answer': answer
            }
            b = json.dumps(query)
            f.write(b)
            f.write('\n')
    for ans in factor_ans:
        if re.search('No', ans):
            no_with_recognize += 1
            break
print(yes_count)
print(no_count)
print(no_with_recognize)


            

                
            
```
