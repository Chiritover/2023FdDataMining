```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = [[500,314,146],[500,223,225]]
hight = [(i[0]-i[1])/i[0] for i in data]
print(hight)
hight2 = [i[2]/i[0] for i in data]
print(hight2)
```

```python
data_prompt1 = [[43,32],[71,57]]
hight_prompt1 = [i[0]/500 for i in data_prompt1]
print(hight_prompt1)
hasattr_prompt1 = [i[1]/500 for i in data_prompt1]
print(hasattr_prompt1)
```

```python
data_temperature = [[500,309,139],[500,238,209]]
hight_temperature = [(i[0]-i[1])/i[0] for i in data_temperature]
high2_temperature = [i[2]/i[0] for i in data_temperature]
print(hight_temperature)
print(high2_temperature)
data_temperature2 = [[500,281,131],[500,243,204]]
hight_temperature2 = [(i[0]-i[1])/i[0] for i in data_temperature2]
high2_temperature2 = [i[2]/i[0] for i in data_temperature2]
```

```python
plt.figure(figsize=(6,4))
plt.style.use('ggplot')
plt.ylim(0,100)
x = np.arange(2)
width = 0.3
plt.bar(x, [hei*100 for hei in hight], width,color = 'dodgerblue',edgecolor = 'white')
plt.bar(x+width, [hei*100 for hei in hight_temperature],width = width,color = 'lightskyblue',edgecolor = 'white')
plt.bar(x+2*width, [hei*100 for hei in hight_temperature2], width = width,color = 'skyblue',edgecolor = 'white')
# 第二个柱状图为网格线,颜色设置为白色

plt.bar(x, [hei * 100 for hei in hight2], width,hatch = 'xx',color = 'dodgerblue',edgecolor = 'white')
plt.bar(x+width, [hei * 100 for hei in high2_temperature ],width =  width,hatch = 'xx',color = 'lightskyblue',edgecolor = 'white')
plt.bar(x+2*width, [hei * 100 for hei in high2_temperature2],width= width,hatch = 'xx',color = 'skyblue',edgecolor = 'white')

plt.xticks(x+width, ['Primality\nTesting','Senator\nSearch'],fontsize=12)
# set y label
plt.ylabel('Proportion of Dataset(%)',fontsize=12)
# 设置图例
plt.legend(['0.0','0.6','0.9'],fontsize=7)
plt.savefig('./result_fig/fig3.png',dpi=300,bbox_inches='tight')
```

```python
plt.figure(figsize=(4,4))
plt.style.use('ggplot')
plt.ylim(0,100)
x = np.arange(2)
width = 0.3
plt.bar(x, [hei*100 for hei in hight], width,color = 'dodgerblue',edgecolor = 'white')
plt.bar(x+width, [hei*100 for hei in hight_prompt1], width,color = 'mediumslateblue',edgecolor = 'white')
# 第二个柱状图为网格线,颜色设置为白色

plt.bar(x, [hei * 100 for hei in hight2], width,hatch = 'xx',color = 'dodgerblue',edgecolor = 'white')
plt.bar(x+width, [hei * 100 for hei in hasattr_prompt1], width,hatch = 'xx',color = 'mediumslateblue',edgecolor = 'white')

plt.xticks(x+0.15, ['Primality\nTesting','Senator\nSearch'],fontsize=12)
# set y label
plt.ylabel('Proportion of Dataset(%)',fontsize=12)
# 设置图例
plt.legend(['Original','Step-by-step prompt'],fontsize=7)
plt.savefig('./result_fig/fig2.png',dpi=300,bbox_inches='tight')
```

```python
plt.figure(figsize=(4,4))
plt.style.use('ggplot')
plt.ylim(0,100)
x = np.arange(2)
width = 0.6
plt.bar(x, [hei*100 for hei in hight], width,color = 'dodgerblue',edgecolor = 'white')
# 第二个柱状图为网格线,颜色设置为白色

plt.bar(x, [hei * 100 for hei in hight2], width,hatch = 'xx',color = 'dodgerblue',edgecolor = 'white')
plt.xticks(x, ['Primality\nTesting','Senator\nSearch'],fontsize=12)
# set y label
plt.ylabel('Proportion of Dataset(%)',fontsize=12)
# 设置图例
plt.legend(['Wrong answer (Hallucination)','Model knows it\'s wrong (Snowballed Hallucination)'],fontsize=7)
plt.savefig('./result_fig/fig1.png',dpi=300,bbox_inches='tight')
```
